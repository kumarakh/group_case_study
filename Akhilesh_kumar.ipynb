## Data understanding

#### Read loan data

```python
import pandas as pd
loan_df = pd.read_csv("loan.csv",low_memory=False)
print(loan_df.head(5))
```

#### Count total rows and column

```python

count_row = loan_df.shape[0]  # number of row count
count_col = loan_df.shape[1]  # number of col count
print(count_row)
print(count_col)
```

#### count number of null/empty rows

```python

count_empty = loan_df.isnull().sum(axis=0)
print(count_empty.count())
```

## Data Cleaning

#### Remove coulmns with all null

```python
loan_final_df = loan_df.dropna(axis='columns', how='all')
filtered_non_empty = loan_final_df.isnull().sum(axis=0)
print(filtered_non_empty.count())
```

```python
loan_final_df.head(5)
```

#### Calculate number of null/empty %

```python
print(loan_final_df.isna().mean().round(4) * 100)
```

#### Remove columnns with more than 90% null

```python
loan_final_df = loan_final_df.loc[:,loan_final_df.isna().mean().round(4) * 100 < 90.00]
loan_final_df.head(5)
```
